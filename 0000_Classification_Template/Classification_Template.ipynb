```python
import pandas as pd

df = pd.read_csv("./classification.csv")

# If you want to remove some columns from the data in advance
# df = df.drop("column name", axis = 1)

# If you want to convert a categorical variable into multiple columns
# you can do it with the following code:
# df = pd.get_dummies(df, columns = ["columnname"])

df.head()
```

```python
from sklearn.model_selection import train_test_split

# Which columns to use for prediction
X = df[["age", "interest"]].values

# Or: The column "success" should not be used for prediction:
# X = df.drop("success", axis = 1).values

y = df["success"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```

```python
# Note: For this command to work, the 
# "helper.py" file must be in the same folder as the current Jupyter notebookok

from helper import plot_classifier

# Plot training data
plot_classifier(model, X_train, y_train, proba = False, xlabel = "Alter", ylabel = "Interesse")
```

```python
# Plot test data

plot_classifier(model, X_test, y_test, proba = False, xlabel = "Alter", ylabel = "Interesse")
```
